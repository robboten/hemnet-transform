```python
df_dropped=df.drop(
    [
        'project',
        'new_construction',
        'iconName',
        'land_area',
        'user_saved',
        'save_button_label',
        'broker_logo_url',
        'broker_name',
        'broker_phone',
        'broker_email',
        'age_string',
        'show_visible_project_units',
        'should_display_showings',
        'upcoming'
    ],
    axis=1)
df_dropped
```

```python
#get the labels column to work with
series_temp=df_dropped[['labels','id']]
series_temp
```

```python

#make series to json then back to df again
js=series_temp.to_json(orient='records')
d1=pd.read_json(js)

d1
```

```python

d2=pd.json_normalize(d1['labels'])
d2
```

```python
d2.shape
```

```python
#list with frames containing each column in expanded df
lista=[]
for i in range(0,d2.shape[1]):
    lista.append(pd.json_normalize(d2[i]))
    #print(dict(lista[i])['identifier'].value_counts())

#lista[0]['identifier'].value_counts()
lista[2].value_counts()
```

```python
#loop thru the features for all columns and add it all to an array, then logical or all bools into one
import numpy

#how to get a list of all identifiers in a category feature?
feature_list=['balcony','elevator','patio','tenure_owned']
compact_df_list=[]

for f in feature_list:
    series=[]
    for d in lista:
        series.append(np.where((d['identifier']==f),True,False))
    narr=numpy.array(series)
    reducedarr=np.logical_or.reduce((narr))
    compact_df_list.append(pd.DataFrame(reducedarr,columns=[f]))

```

```python
#expand the livingspace.table
series_livingspace=df_dropped['living_space.table']
df_livingspace=pd.json_normalize(series_livingspace)
#df_livingspace.reset_index()
```

d1.reset_index(drop=True)


df_livingspace.reset_index(drop=True)

```python
features=pd.concat(compact_df_list,axis=1)
features
```

df_dropped.reset_index(drop=True).head(2)

```python

merged_dfs=pd.concat([features,df_livingspace,d1],axis=1)
merged_dfs
```

```python
#merge it all
dfs = [df_dropped.set_index('id'),merged_dfs.set_index('id')]
all=pd.concat(dfs, axis=1)
all.drop(['labels','living_space.table'],axis=1,inplace=True)
```
